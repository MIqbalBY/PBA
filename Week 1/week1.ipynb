## 📚 Bagian 1: Analisis Teks Bahasa Indonesia

Pada bagian ini, kita akan menganalisis dataset sentimen bahasa Indonesia (IndoNLU SMSA) dengan melakukan:
- Tokenisasi menggunakan NLTK
- Stemming menggunakan Sastrawi
- Analisis frekuensi kata

### 1.1 Import Library dan Persiapan Data

```python
# Install packages yang diperlukan
%pip install datasets nltk PySastrawi -q
```

```python
# Import semua library yang dibutuhkan
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from collections import Counter

# Download resource NLTK untuk tokenisasi
nltk.download('punkt', quiet=True)

print("✅ Library berhasil di-import!")
```

### 1.2 Load Dataset SMSA (Sentiment Analysis)

Dataset ini berisi review dan sentimen dalam bahasa Indonesia dari IndoNLU benchmark.

```python
print("📥 Mengunduh dataset SMSA dari GitHub IndoNLU...")

# Load dataset langsung dari GitHub (bypass Hugging Face)
url = "https://raw.githubusercontent.com/indobenchmark/indonlu/master/dataset/smsa_doc-sentiment-prosa/train_preprocess.tsv"
df = pd.read_csv(url, sep='\t', header=None, names=['text', 'label'])

print(f"✅ Berhasil memuat {len(df):,} baris data!")
print(f"\nSample data:")
print(df.head())
```

### 1.3 Pilih Sample untuk Analisis

```python
# Ambil 3 sampel pertama untuk analisis detail
sample_texts = df['text'][:3].tolist()  
labels = df['label'][:3].tolist()

print("📋 Sample teks yang akan dianalisis:")
for i, (text, label) in enumerate(zip(sample_texts, labels), 1):
    print(f"\n{i}. Label: {label}")
    print(f"   Teks: {text[:100]}..." if len(text) > 100 else f"   Teks: {text}")
```

### 1.4 Tokenisasi dan Stemming

**Tokenisasi**: Memecah teks menjadi unit-unit kecil (token)  
**Stemming**: Mengubah kata ke bentuk dasarnya (contoh: "berlari" → "lari")

```python
# Inisialisasi stemmer Sastrawi
stemmer = StemmerFactory().create_stemmer()

# Proses setiap teks
for i, text in enumerate(sample_texts):
    print("="*80)
    print(f"📝 ANALISIS TEKS {i+1}")
    print(f"Label Sentimen: {labels[i]}")
    print("="*80)
    
    print(f"\n📄 Teks Asli:")
    print(f"   {text}")
    
    # Tokenisasi
    tokens = word_tokenize(text)
    print(f"\n🔤 Hasil Tokenisasi:")
    print(f"   {tokens}")
    print(f"\n📊 Statistik Token:")
    print(f"   • Total token: {len(tokens)}")
    print(f"   • Token unik: {len(set(tokens))}")
    
    # Stemming
    stemmed_text = stemmer.stem(text)
    stemmed_tokens = word_tokenize(stemmed_text)
    print(f"\n🌱 Hasil Stemming:")
    print(f"   {stemmed_tokens}")
    print(f"\n📊 Statistik Setelah Stemming:")
    print(f"   • Total token: {len(stemmed_tokens)}")
    print(f"   • Token unik: {len(set(stemmed_tokens))}")
    
    # Simpan token ke file
    with open(f'tokens_sample_{i+1}.txt', 'w', encoding='utf-8') as f:
        f.write('\n'.join(tokens))
    print(f"\n💾 Token disimpan ke: tokens_sample_{i+1}.txt")
    print()
```

### 1.5 Analisis Frekuensi Kata

Mari kita lihat kata-kata yang paling sering muncul dalam sample teks.

```python
# Hitung frekuensi kata dari semua sample
all_tokens = []
for text in sample_texts:
    all_tokens.extend(word_tokenize(text.lower()))

word_freq = Counter(all_tokens).most_common(10)

print("📊 10 KATA TERATAS (Berdasarkan Frekuensi)")
print("="*50)
for rank, (word, count) in enumerate(word_freq, 1):
    bar = "█" * count
    print(f"{rank:2}. {word:15} | {count:2}x | {bar}")
```

## 📖 Bagian 2: Analisis Teks Bahasa Inggris

Pada bagian ini, kita akan menganalisis novel klasik "Emma" karya Jane Austen dari corpus Gutenberg.

### 2.1 Load Corpus Gutenberg

```python
# Download corpus Gutenberg
nltk.download('gutenberg', quiet=True)
nltk.download('punkt_tab', quiet=True)

print("✅ Corpus Gutenberg siap digunakan!")
```

### 2.2 Tokenisasi Novel "Emma"

```python
# Load teks dari corpus Gutenberg
print("📚 Memuat novel 'Emma' by Jane Austen...")
text = nltk.corpus.gutenberg.raw('austen-emma.txt')

print(f"✅ Novel berhasil dimuat!")
print(f"\n📄 Preview 500 karakter pertama:")
print("-" * 80)
print(text[:500])
print("-" * 80)
```

### 2.3 Analisis Token

```python
# Tokenisasi
print("🔤 Melakukan tokenisasi...")
tokens = word_tokenize(text)

# Tampilkan hasil
print(f"\n✨ Hasil Tokenisasi 'Emma':")
print("="*80)
print(f"\n📝 20 Token Pertama:")
print(tokens[:20])

print(f"\n📊 STATISTIK DOKUMEN:")
print("-" * 50)
print(f"  • Total tokens     : {len(tokens):,}")
print(f"  • Token unik       : {len(set(tokens)):,}")
print(f"  • Rasio keunikan   : {len(set(tokens))/len(tokens)*100:.2f}%")
print(f"  • Panjang teks     : {len(text):,} karakter")
print("="*80)
```
