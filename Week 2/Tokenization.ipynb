## 1️⃣ White Space Tokenization

**Definisi:**  
White Space Tokenization adalah teknik tokenisasi paling sederhana yang memisahkan teks berdasarkan spasi (whitespace).

**Karakteristik:**
- ✅ Sangat sederhana dan cepat
- ✅ Menggunakan fungsi `.split()` bawaan Python
- ⚠️ Tidak menangani tanda baca dengan baik
- ⚠️ Tidak cocok untuk teks kompleks

**Library yang digunakan:** Python built-in, NLTK

### 1.1. Menggunakan Python Built-in `.split()`

```python
# Contoh teks dalam Bahasa Indonesia
kalimat = "Dikutip dari Reuters, dia menyampaikan permintaan maaf sebelum mengundurkan diri karena telah berlibur dengan keluarganya selama 4 minggu tak lama setelah banjir terjadi dan menewaskan lebih dari 100 orang."

# Tokenisasi menggunakan .split()
tokens = kalimat.split()

print("=" * 80)
print("📌 TEKS ASLI:")
print(kalimat)
print("\n" + "=" * 80)
print("🔹 HASIL TOKENISASI:")
print(tokens)
print("\n" + "=" * 80)
print(f"📊 JUMLAH TOKEN: {len(tokens)} tokens")
print("=" * 80)
```

### 1.2. Menggunakan NLTK WhitespaceTokenizer

```python
# Menggunakan WhitespaceTokenizer dari NLTK
wtk = WhitespaceTokenizer()

teks_contoh = "Computers offer powerful capabilities for searching and reasoning about structured records and relational data"

# Tokenisasi
tokens_nltk = wtk.tokenize(teks_contoh)

print("=" * 80)
print("📌 TEKS ASLI:")
print(teks_contoh)
print("\n" + "=" * 80)
print("🔹 HASIL TOKENISASI (NLTK WhitespaceTokenizer):")
print(tokens_nltk)
print("\n" + "=" * 80)
print(f"📊 JUMLAH TOKEN: {len(tokens_nltk)} tokens")
print("=" * 80)
```

---

## 2️⃣ Word Tokenization (Tokenisasi Kata)

**Definisi:**  
Word Tokenization memisahkan teks menjadi unit-unit kata individual dengan mempertimbangkan tanda baca dan konteks.

**Karakteristik:**
- ✅ Menangani tanda baca dengan baik
- ✅ Mempertimbangkan konteks bahasa
- ✅ Cocok untuk analisis NLP tingkat lanjut
- ⚠️ Lebih lambat dari white space tokenization

**Library yang digunakan:** NLTK, SpaCy

### 2.1. Word Tokenization menggunakan SpaCy

```python
# Teks sampel
teks = "Menteri Urusan Keluarga di Jerman Anne Spiegel mengundurkan diri setelah kontroversinya berlibur pascabanjir dahsyat melanda Jerman pada 2021"

# Load SpaCy model dengan fallback
try:
    nlp = spacy.load('en_core_web_sm')
except OSError:
    print("⚠️  Model 'en_core_web_sm' tidak ditemukan. Menggunakan tokenizer dasar.")
    nlp = English()

# Proses tokenisasi
doc = nlp(teks)

print("=" * 80)
print("📌 TEKS ASLI:")
print(teks)
print("\n" + "=" * 80)
print("🔹 HASIL TOKENISASI DENGAN INDEKS:\n")

for token in doc:
    print(f"   Token: '{token.text:20s}' | Posisi Indeks: {token.idx}")

print("\n" + "=" * 80)

# Ekstrak token sebagai list
tokens_spacy = [token.text for token in doc]
print("🔹 TOKEN DALAM BENTUK LIST:")
print(tokens_spacy)
print("\n" + "=" * 80)
print(f"📊 JUMLAH TOKEN: {len(tokens_spacy)} tokens")
print("=" * 80)
```

### 2.2. Word Tokenization menggunakan NLTK

```python
# Menggunakan NLTK word_tokenize
tokens_nltk = word_tokenize(teks)

print("=" * 80)
print("📌 TEKS ASLI:")
print(teks)
print("\n" + "=" * 80)
print("🔹 HASIL TOKENISASI (NLTK):")
print(tokens_nltk)
print("\n" + "=" * 80)
print(f"📊 JUMLAH TOKEN: {len(tokens_nltk)} tokens")
print("=" * 80)

# Perbandingan SpaCy vs NLTK
print("\n" + "=" * 80)
print("📊 PERBANDINGAN SpaCy vs NLTK:")
print("=" * 80)
print(f"   SpaCy: {len(tokens_spacy)} tokens")
print(f"   NLTK:  {len(tokens_nltk)} tokens")
print("=" * 80)
```

## 3️⃣ Sentence Tokenization (Tokenisasi Kalimat)

**Definisi:**  
Sentence Tokenization memisahkan teks menjadi kalimat-kalimat individual.

**Karakteristik:**
- ✅ Memahami struktur kalimat 
- ✅ Menangani singkatan (Dr., Mr., dll)
- ✅ Penting untuk analisis dokumen panjang
- ✅ Preprocessing untuk word tokenization

**Library yang digunakan:** NLTK

```python
# Teks dengan beberapa kalimat
teks_paragraf = "Good morning everyone. Welcome to the AI Workshop. Dr. Carter and Michael T. Anderson are waiting for you. They'll join you shortly."

print("=" * 80)
print("📌 TEKS ASLI:")
print(teks_paragraf)
print("\n" + "=" * 80)
print("🔹 HASIL TOKENISASI KALIMAT:\n")

# Tokenisasi per kalimat
kalimat_list = sent_tokenize(teks_paragraf)

for i, kalimat in enumerate(kalimat_list, 1):
    print(f"   Kalimat {i}: {kalimat}")

print("\n" + "=" * 80)
print(f"📊 JUMLAH KALIMAT: {len(kalimat_list)} kalimat")
print("=" * 80)

# Tokenisasi kata per kalimat
print("\n" + "=" * 80)
print("🔹 TOKENISASI KATA PER KALIMAT:\n")
for i, kalimat in enumerate(kalimat_list, 1):
    tokens = word_tokenize(kalimat)
    print(f"   Kalimat {i}: {tokens}")
    print(f"   → Jumlah kata: {len(tokens)} tokens\n")
print("=" * 80)
```

## 4️⃣ Character Tokenization (Tokenisasi Karakter)

**Definisi:**  
Character Tokenization memisahkan teks menjadi karakter-karakter individual.

**Karakteristik:**
- ✅ Paling granular (detail)
- ✅ Berguna untuk model character-level
- ✅ Menangani kata OOV (Out of Vocabulary)
- ⚠️ Menghasilkan sequence yang sangat panjang

**Penggunaan:** Character-level RNN, OCR, Spelling correction

```python
# Teks sampel
kata = "Jerman"

# Character tokenization
characters = list(kata)

print("=" * 80)
print("📌 KATA ASLI:")
print(kata)
print("\n" + "=" * 80)
print("🔹 HASIL TOKENISASI KARAKTER:")
print(characters)
print("\n" + "=" * 80)
print(f"📊 JUMLAH KARAKTER: {len(characters)} karakter")
print("=" * 80)

# Contoh dengan spasi
print("\n" + "=" * 80)
print("📌 CONTOH DENGAN KALIMAT LENGKAP:")
kalimat_pendek = "NLP 2026"
chars_kalimat = list(kalimat_pendek)
print(f"Teks: {kalimat_pendek}")
print(f"Tokenisasi: {chars_kalimat}")
print(f"Jumlah: {len(chars_kalimat)} karakter (termasuk spasi)")
print("=" * 80)
```

## 5️⃣ Subword Tokenization (Tokenisasi Sub-Kata)

**Definisi:**  
Subword Tokenization memecah kata menjadi unit-unit yang lebih kecil (subwords) untuk menangani kata-kata yang jarang atau tidak dikenal.

**Karakteristik:**
- ✅ Menangani kata OOV (Out of Vocabulary)
- ✅ Efisien untuk model deep learning
- ✅ Balance antara character dan word level
- ✅ Digunakan di BERT, GPT, dll

**Contoh:**  
`"urbanization"` → `["urban", "##ization"]`

**Library yang digunakan:** Transformers (BERT Tokenizer)

```python
# Load BERT Tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Contoh teks 1
teks1 = "Public opposition to the urbanization of the countryside"
subwords1 = tokenizer.tokenize(teks1)

# Contoh teks 2
teks2 = "Tokenization is essential in NLP"
subwords2 = tokenizer.tokenize(teks2)

print("=" * 80)
print("📌 CONTOH 1:")
print(f"Teks Asli: {teks1}")
print(f"\n🔹 Subword Tokens:")
print(subwords1)
print(f"\n💡 Perhatikan: 'urbanization' → {['urban', '##ization']}")
print("\n" + "=" * 80)

print("📌 CONTOH 2:")
print(f"Teks Asli: {teks2}")
print(f"\n🔹 Subword Tokens:")
print(subwords2)
print(f"\n💡 Perhatikan: 'nlp' → {['nl', '##p']}")
print("\n" + "=" * 80)

print("📊 STATISTIK:")
print(f"   Contoh 1: {len(teks1.split())} kata → {len(subwords1)} subwords")
print(f"   Contoh 2: {len(teks2.split())} kata → {len(subwords2)} subwords")
print("=" * 80)

print("\n💡 KETERANGAN:")
print("   Prefix '##' menunjukkan bahwa token tersebut adalah lanjutan dari token sebelumnya")
print("=" * 80)
```

## 6️⃣ Perbandingan Metode Tokenisasi Tingkat Lanjut

Pada bagian ini, kita akan membandingkan berbagai metode tokenisasi yang lebih canggih:

1. **White Space** - Tokenisasi sederhana dengan spasi
2. **Regular Expression (Regex)** - Pattern matching untuk token
3. **Penn Treebank** - Aturan linguistik untuk parsing
4. **SpaCy** - Tokenizer berbasis machine learning
5. **Moses** - Tokenizer untuk machine translation

Mari kita lihat perbedaannya pada teks yang sama!

```python
# Teks sampel untuk perbandingan
teks_perbandingan = "Pengunduran diri tersebut dilakukan atas keputusannya sendiri pada 11 April 2022. Spiegel diketahui berlibur setelah bencana banjir terjadi di negara bagian tempat dia menjabat sebagai pejabat senior."

# Definisi fungsi-fungsi tokenisasi

def whitespace_tokenization(text):
    """Tokenisasi sederhana berdasarkan spasi"""
    return text.split()

def regex_tokenization(text):
    """Tokenisasi menggunakan Regular Expression"""
    tokenizer = RegexpTokenizer(r'\w+')
    return tokenizer.tokenize(text)

def penn_treebank_tokenization(text):
    """Tokenisasi dengan aturan Penn Treebank"""
    tokenizer = TreebankWordTokenizer()
    return tokenizer.tokenize(text)

def spacy_tokenization(text):
    """Tokenisasi menggunakan SpaCy"""
    try:
        nlp_model = spacy.load("en_core_web_sm")
    except:
        nlp_model = English()
    doc = nlp_model(text)
    return [token.text for token in doc]

def moses_tokenization(text):
    """Tokenisasi menggunakan Moses"""
    moses_tok = MosesTokenizer()
    return moses_tok.tokenize(text)

print("✅ Fungsi tokenisasi telah didefinisikan!")
```

```python
# Jalankan semua metode tokenisasi
hasil = {
    "White Space": whitespace_tokenization(teks_perbandingan),
    "Regex": regex_tokenization(teks_perbandingan),
    "Penn Treebank": penn_treebank_tokenization(teks_perbandingan),
    "SpaCy": spacy_tokenization(teks_perbandingan),
    "Moses": moses_tokenization(teks_perbandingan)
}

# Tampilkan hasil dengan format rapi
print("=" * 80)
print("📌 TEKS ASLI:")
print(teks_perbandingan)
print("\n" + "=" * 80)
print("🔍 PERBANDINGAN HASIL TOKENISASI:\n")

for metode, tokens in hasil.items():
    print(f"📊 {metode}:")
    print(f"   Jumlah token: {len(tokens)}")
    print(f"   Hasil: {tokens[:10]}...")  # Tampilkan 10 token pertama
    print()

print("=" * 80)
print("📊 RINGKASAN JUMLAH TOKEN:")
print("=" * 80)
for metode, tokens in hasil.items():
    print(f"   {metode:20s}: {len(tokens):3d} tokens")
print("=" * 80)

# Analisis perbedaan
print("\n" + "=" * 80)
print("💡 ANALISIS PERBEDAAN:")
print("=" * 80)
print("1. White Space: Paling sederhana, memisahkan berdasarkan spasi saja")
print("2. Regex: Menghilangkan tanda baca, fokus pada kata alphanumeric")
print("3. Penn Treebank: Memisahkan tanda baca menjadi token terpisah")
print("4. SpaCy: ML-based, memahami konteks bahasa")  
print("5. Moses: Dirancang khusus untuk machine translation")
print("=" * 80)
```

## 📝 Kesimpulan

Dalam notebook ini, kita telah mempelajari berbagai teknik tokenisasi dalam NLP:

### 🎯 Poin Penting:

1. **White Space Tokenization** - Sederhana tapi terbatas
2. **Word Tokenization** - Standar untuk NLP, menangani tanda baca dengan baik
3. **Sentence Tokenization** - Penting untuk analisis dokumen
4. **Character Tokenization** - Granular, untuk model character-level
5. **Subword Tokenization** - Modern approach, mengatasi masalah OOV

### 💡 Kapan Menggunakan Masing-Masing?

| Metode | Kasus Penggunaan |
|--------|------------------|
| White Space | Preprocessing sederhana, data bersih |
| Word | Kebanyakan task NLP (klasifikasi, NER, dll) |
| Sentence | Document summarization, text generation |
| Character | OCR, spell checking, morphological analysis |
| Subword | Transformer models (BERT, GPT), multilingual |

### 🔧 Library Rekomendasi:

- **NLTK**: Baik untuk pembelajaran, banyak resource
- **SpaCy**: Production-ready, cepat, akurat
